## This Jupyter Python 2.7 notebook searches ADS for papers related to your favourite keyword, author or whatever you like really, and generates abstracts that look like the results using a Markov model.

```python
query = '^pope,benjamin j. s.'
query = 'exoplanet'
query 

depth = 1

name = 'test'
```

```python
papers = ads.SearchQuery(q=query,max_pages=1000, sort="citation_count")

# abstracts = ''
# titles = ''
titles = []
# titles = [paper.title[0].encode('ascii','ignore') for paper in papers]
for paper in papers:
    try: 
        thistitle = paper.title[0].encode('ascii','ignore') 
        print thistitle
        titles.append(thistitle)
    except:
        pass
```

```python
delimiter = '. '

corpus = delimiter.join(titles)
```

```python
len(corpus)
```

```python
### If you haven't made your database already, do this! It may take some time and lots of disk space.

os.system('rm %s.db' % name)
db = Db(sqlite3.connect(name + '.db'), Sql())
depth = 3
db.setup(depth)

Parser(name, db, SENTENCE_SEPARATOR, WORD_SEPARATOR).parse(corpus)
```

```python
### generate some text

db = Db(sqlite3.connect(name + '.db'), Sql())
generator = Generator(name, db, Rnd())
count = 5
newtext = ''

for i in range(0, count):
    while newtext in corpus:
        newtext = generator.generate(WORD_SEPARATOR)#+'. '
    
print newtext
```
